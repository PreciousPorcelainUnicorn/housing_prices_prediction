```python
df = pd.read_csv('/kaggle/input/home-data-for-ml-course/train.csv')
df.head()
```

```python
df.dropna(subset=['Electrical'], inplace=True)
```

```python
df.drop(columns=['MSZoning',
 'LotFrontage',
 'Alley',
 'Utilities',
 'Exterior1st',
 'Exterior2nd',
 'MasVnrType',
 'MasVnrArea',
 'BsmtQual',
 'BsmtCond',
 'BsmtExposure',
 'BsmtFinType1',
 'BsmtFinSF1',
 'BsmtFinType2',
 'BsmtFinSF2',
 'BsmtUnfSF',
 'TotalBsmtSF',
 'BsmtFullBath',
 'BsmtHalfBath',
 'KitchenQual',
 'Functional',
 'FireplaceQu',
 'GarageType',
 'GarageYrBlt',
 'GarageFinish',
 'GarageCars',
 'GarageArea',
 'GarageQual',
 'GarageCond',
 'PoolQC',
 'Fence',
 'MiscFeature',
 'SaleType'], axis=1,inplace=True)
```

```python
df.drop(columns='MSSubClass', axis=1, inplace=True)
```

```python
neigh = {'Blmngtn':'Bloomington Heights', 'Blueste':'Bluestem', 'BrDale':'Briardale', 'BrkSide':'Brookside', 'ClearCr':'Clear Creek', 'CollgCr':'College Creek',
         'Crawfor':'Crawford', 'Edwards':'Edwards', 'Gilbert':'Gilbert', 'IDOTRR':'Iowa DOT and Rail Road', 'MeadowV':'Meadow Village',
         'Mitchel':'Mitchell', 'NAmes':'North Ames', 'NoRidge':'Northridge','NPkVill':'Northpark Villa', 'NridgHt':'Northridge Heights', 'NWAmes':'Northwest Ames',
         'OldTown':'Old Town', 'SWISU':'South & West of Iowa State University','Sawyer':'Sawyer','SawyerW':'Sawyer West','Somerst':'Somerset', 
         'StoneBr':'Stone Brook','Timber':'Timberland','Veenker':'Veenker'}
df['Neighborhood Full Name'] = df['Neighborhood'].map(neigh)
df.info()
```

```python
df[df['Neighborhood Full Name'].isna()]
```

```python
bldg = {'1Fam':'Single-family Detached', '2fmCon':'Two-family Conversion','Duplex':'Duplex','TwnhsE':'Townhouse End Unit','Twnhs':'Townhouse Inside Unit'}
df['Type of dwelling'] = df['BldgType'].map(bldg)
df.info()
```

```python
df[df['Type of dwelling'].isna()]
```

```python
df['BldgType'].unique()
```

```python
df['HouseStyle'].unique()
```

```python
hstyle = {'2Story':'Two story', '1Story': 'One story', '1.5Fin':'One and one-half story: 2nd level finished', '1.5Unf':'One and one-half story: 2nd level unfinished', 'SFoyer':'Split Foyer', 'SLvl':'Split Level', '2.5Unf':'Two and one-half story: 2nd level unfinished',
       '2.5Fin':'Two and one-half story: 2nd level finished'}
df['HouseStyle'] = df['HouseStyle'].map(hstyle)
df.info()
```

```python
df['Foundation'].unique()
```

```python
found = {'PConc':'Poured Contrete', 'CBlock':'Cinder Block', 'BrkTil':'Brick & Tile', 'Wood':'Wood', 'Slab':'Slab', 'Stone':'Stone'}
df['Foundation'] = df['Foundation'].map(found)
df.info()
```

```python
df.isna().sum()
```

```python
df['RoofStyle'].unique()
```

```python
rstyle = {'Gable':'Gable', 'Hip':'Hip', 'Gambrel':'Gambrel (Barn)', 'Mansard':'Mansard', 'Flat':'Flat', 'Shed':'Shed'}
df['RoofStyle'] = df['RoofStyle'].map(rstyle)
df.info()
```

```python
df['RoofMatl'].unique()
```

```python
rmat = {'CompShg':'Standard (Composite) Shingle', 'WdShngl':'Wood Shingles', 'Metal':'Metal', 'WdShake':'Wood Shakes', 'Membran':'Membrane', 'Tar&Grv':'Gravel & Tar',
       'Roll':'Roll', 'ClyTile':'Clay or Tile'}
df['Roof Material'] = df['RoofMatl'].map(rmat)
df.info()
```

```python
df['Heating'].unique()
```

```python
heat = {'GasA':'Gas forced warm air furnace', 'GasW':'Gas hot water or steam heat', 'Grav':'Gravity furnace', 'Wall':'Wall furnace', 'OthW':'Hot water or steam heat other than gas', 'Floor':'Floor Furnace'}
df['Heating'] = df['Heating'].map(heat)
df.info()
```

```python
df['HeatingQC'].unique()
```

```python
heatc = {'Ex':'Excellent', 'Gd':'Good', 'TA':'Average/Typical', 'Fa':'Fair', 'Po':'Poor'}
df['HeatingQC'] = df['HeatingQC'].map(heatc)
df.info()
```

```python
df['SaleCondition'].unique()
```

```python
scond = {'Normal':'Normal Sale', 'Abnorml':'Abnormal Sale -  trade, foreclosure, short sale', 'Partial':'Home was not completed when last assessed (associated with New Homes)', 'AdjLand':'Adjoining Land Purchase', 'Alloca':'Allocation - two linked properties with separate deeds, typically condo with a garage unit', 'Family':'Sale between family members'}
df['SaleCondition'] = df['SaleCondition'].map(scond)
df.info()
```

```python
df.rename(columns={'LotArea':'Lot size (square feet)', 'OverallQual':'Overall quality of material and finish', 'OverallCond':'Overall condition', 
                   'YearBuilt':'Original construction year', 'YearRemodAdd':'Remodel year', 'RoofStyle': 'Type of roof', 'RoofMatl':'Roof material', 'Foundation':'Type of foundation',
                   'HeatingQC': 'Heating quality and condition', '1stFlrSF':'First Floor square feet', '2ndFlrSF':'Second floor square feet', 'LowQualFinSF':'Low quality finished square feet (all floors)', 
                   'GrLivArea':'Above grade (ground) living area (square feet)', 'FullBath':'Full bathrooms above grade', 'TotRmsAbvGrd':'Total rooms above grade (does not include bathrooms)', 
                   'Fireplaces': 'Number of fireplaces', 'OpenPorchSF':'Open porch area in square feet', 'EnclosedPorch':'Enclosed porch area in square feet',
                    'YrSold':'Year Sold', 'SaleCondition':'Condition of sale'}, inplace=True)
df.head()
```
